```python
# password validation

def password_validation(password):
    lower = 0
    upper = 0
    digit = 0
    special = 0
    space = 0
    for i in password:
        if i.islower():
            lower+=1
            
        elif i.isupper():
            upper+=1
            
        elif i.isdigit():
            digit +=1
            
        elif  i.isspace():
            space +=1
            
        elif  not i.isidentifier():
            special +=1
            
       
            
#     print(space)
#     print(special)
            
    if lower>=1 and upper>=1 and digit>=1 and special>=1 and space ==0 and len(password)>=8:
        return True
    else:
        return False
    
    
# sighup

def signup():
    user_name = input("Enter name: ")
    password = getpass("Enter password: ")
    security_code = getpass("Enter security key: ")
    
    if password_validation(password):
        print("Initial amount of 5000/- is required to active your account ")
        print("is you are agree press yes(y) otherwise no(n):")
        
        choice = input("y/n").lower()
        

    
    
        
        if choice == "y":
            print("we will provide you a uniuqe acc number:")
            acc_no  = list(data.keys())[-1] + 1
            balance = 5000
            data[acc_no] = [user_name ,password , balance , security_code]
            print("Account Created successfilly.....")
            print("your account number is  " , acc_no)
            
            
        
        elif choice == "n":
             print("okay")
           
        
        else:
            print("please select yes or no ")
            
    else:
        print("Please enter valid password") 
        
#login

def login():
    acc_no = int(input("ENTER YOUR ACCOUNT NUMBER: "))
    password = getpass("ENTER PASSWORD : ")
    
    if acc_no in data.keys() and password == data[acc_no][1]:
        print("login successfilly.....")
        while True :
            print("1) cradit")
            print("2) Dabit")
            print("3) Change password")
            print("4) Check balance ")
            print("5) Show detalis")
            print("6) logout")
            
            choise  = int(input())
            if choise == 1:
                cradit_amount = float(input("Enter amount to cradit : "))
                data[acc_no][2] += cradit_amount
                print("cradited succeessfully...")
                print("total amount : ", data[acc_no][2])
                
            elif choise ==2:
                dabit_amount = float(input("Enter amount to dabit: "))
                if (data[acc_no][2] - dabit_amount)>=5000:
                    data[acc_no][2]-=dabit_amount
                    print(f"{dabit_amount} dabit successfully")
                    print("total amount : ", data[acc_no][2])
                    
                else:
                    print("no enough balance")
                    
            elif choise ==3:
                old_password = getpass("Enter old password: ")
                if old_password == data[acc_no][1]:
                    new_password = getpass("enter new password: ")
                    if password_validation(new_password):
                        data[acc_no][1] = new_password
                        print("Changed successfully...")
                        
                    else:
                        print("please enter valid password")
                   
                else:
                    print("Your old password is incorrect")
                    
            elif choise ==4:
                print(f"your account balance is {data[acc_no][2]}")
                
            elif choise ==5:
                print(f"account number : {[acc_no]} \nuser name : {data[acc_no][0]} \nbalance : {data[acc_no][2]}")
                
            elif choise == 6:
                break
                
                
            else:
                print("valid............")
                
          
    else:
        print("invalid password  try again")
        
# forget password

def forget_password():
#     print("if Your are forget Your password than u have to visit branch ")
    acc_no = int(input("Enter Your account number "))
    user_name = input("Enter user name")
    security_code = getpass("Enter security key ")
    if acc_no in data.keys() and user_name  == data[acc_no][0] and security_code == data[acc_no][3] :
        new_password = getpass("Enter your new password")
        if password_validation(new_password):
            print("password recovered successfully...")
            data[acc_no][1] = new_password 
            
        else:
            print("please enter vlaid password")
        
    else:
        print("please enter valid details ....\ntry again")
        
# home

def home():
    print("1) login \n2)signup")
    
    
    home_choise = int(input())
    if home_choise == 1:
        print("1) login \n2)forget password")
        log_c = int(input())
        if log_c == 1:
            login()
            
        elif log_c == 2:
            forget_password()
            
        else:
            print("enter valid number ")
            
    elif home_choise == 2:
        signup()
        
    else:
         print("enter valid number ")
        

        
```
